```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in Obj and DC data from AMPL Relax model solutions
Obj = np.loadtxt("RelaxData/objectRelax.txt", delimiter = "\n")
DC = np.loadtxt("RelaxData/DCRelax.txt", delimiter = "\n")
```

```python
# Read in X and Z solutions from AMPL Relax model sols
X = pd.read_csv('RelaxData/XRelaxCleaned.txt', sep = ' ', header = None)
Z = pd.read_csv('RelaxData/ZRelaxCleaned.txt', sep = ' ', header = None)
```

```python
# Drop extra null columns
X = X.drop(480, axis = 1)
Z = Z.drop(114960, axis = 1)
```

```python
# Redo column indexing due to how data was cleaned
X.columns = [x + 48 * y for x in range(0, 48) for y in range(0, 10)]
Zcols = [x + 16423 * y for x in range(0, 16423) for y in range(0, 7)]
# Have to get rid of last entry of Zcols
Zcols.pop()
Z.columns = Zcols
```

```python
# Add DC, Obj columns
X['DC'] = DC
X['Obj'] = Obj
Z['DC'] = DC
Z['Obj'] = Obj
```

```python
# Read in points data
pointsData = pd.read_csv('PointsData.csv')
# Just need X and Y columns
pointsData = pointsData.iloc[:,1:3]
pointsData.tail()
```

# X Code:
  This code creates a dataframe that has every Radar point for the rows.
  The columns are the X and Y coordinates (in miles), DC, and Obj value

```python
# for every row,
#   oneEntries <- python list (vector) of column indexes where entry == 1  (length should == 1)
#   Subset the ``points'' dataframe to get X and Y values for those points
#   X, Y, DC, Obj, # Radars (length of oneEntries) gets row binded onto new dataframe
#      X.loc[:, 'Obj'].sum() shows should be 207 points -> num rows in new dataframe
#   pd.concat:   axis = 0 is rbind, axis = 1 is cbind

allXData = pd.DataFrame()
for rowIndex in range(X.shape[0]):
    oneEntries = (X.iloc[rowIndex,:] == 1)
    oneEntries = oneEntries[oneEntries].index
    dc = X.iloc[rowIndex, 480]
    obj = X.iloc[rowIndex, 481]
    DcObj = pd.concat([pd.Series(np.repeat(dc, 1)), pd.Series(np.repeat(obj, 1))], axis = 1, ignore_index = True)
    data = pd.concat([pointsData.iloc[oneEntries,:].reset_index(drop=True), DcObj], 
                     axis = 1, ignore_index = True)
    data.columns = ['X', 'Y', 'DC', 'Objective']
    allXData = pd.concat([allXData, data], axis = 0)
    allXData.reset_index(drop=True, inplace = True)
    
print(allXData.shape)
print(allXData.head())
allXData.tail() 
```

## Adapting code for Z:
   Change Z columns to be tuple of the points
   
   Get column tuples where Z row == 1  ->   [(), (), (), ...]
   
   Get unique values from previous list and counts of num times they occur (like table() in R)
   
   Make rows of dataframe with fields:
   
        X     Y    Num Not Resolving    DC   Objective

```python
# Get numpy array of possible combinations of points
from itertools import combinations
nodePairs = np.array(list(combinations(range(480), 2)))
```

```python
# for every row,
#   oneEntries <- python list (vector) of column indexes where entry == 1  (length should == Obj)
#     Change column indexes in oneEntries to be pairs of points
#        Get unique points and number of times they each occur (table() in R)
#   Subset the ``points'' dataframe to get X and Y values for those points
#   X, Y, DC, Obj, # Radars (length of oneEntries) gets row binded onto new dataframe
#   pd.concat:   axis = 0 is rbind, axis = 1 is cbind

allZData = pd.DataFrame()
for rowIndex in range(Z.shape[0]):
    oneEntries = (Z.iloc[rowIndex,:] == 1)
    oneEntries = oneEntries[oneEntries].index
    dc = Z.iloc[rowIndex, 114960]
    obj = Z.iloc[rowIndex, 114961]
    
    # Get the pair of point indexes for each column index
    onePointIndexes = nodePairs[list(oneEntries)]
    # Reshape so 1 dimensional vector
    onePointIndexes = onePointIndexes.reshape((onePointIndexes.shape[0] * onePointIndexes.shape[1], 1))
    # Get unique point indexes and number of times each occurs
    pointCounts = np.unique(onePointIndexes, return_counts = True)   # [0] is points, [1] is counts
    
    pointsCountDF = pd.concat([pointsData.iloc[pointCounts[0],:].reset_index(drop=True),
                              pd.Series(pointCounts[1])], axis = 1, ignore_index = True)
    DcObj = pd.concat([pd.Series(np.repeat(dc, len(pointCounts[0]))), 
                       pd.Series(np.repeat(obj, len(pointCounts[0])))], axis = 1, ignore_index = True)
    
    data = pd.concat([pointsCountDF.reset_index(drop=True), DcObj], 
                     axis = 1, ignore_index = True)
    data.columns = ['X', 'Y', '# Times Not Resolving', 'DC', 'Objective']
    allZData = pd.concat([allZData, data], axis = 0)
    allZData.reset_index(drop=True, inplace = True)
    
print(allZData.shape)
print(allZData.head())
allZData.tail()
```
